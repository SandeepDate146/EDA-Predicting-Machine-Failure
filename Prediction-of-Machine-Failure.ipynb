```python
org_data.drop(columns='RNF', inplace=True)

#data_all = pd.concat([df_1, original_1], ignore_index=True)

org_data['air_process_diff'] = abs(org_data['Air temperature [K]'] - org_data['Process temperature [K]'])

org_data['speed_power'] = org_data['Rotational speed [rpm]'] * (2 * np.pi / 60) / (org_data['Rotational speed [rpm]'] * (2 * np.pi / 60) * org_data['Torque [Nm]'])

org_data['torque_power'] = org_data['Torque [Nm]'] / (org_data['Rotational speed [rpm]'] * (2 * np.pi / 60) * org_data['Torque [Nm]'])

org_data["tool_process"]=org_data["Tool wear [min]"] * org_data["Process temperature [K]"]

org_data["temp_ratio"] = org_data["Process temperature [K]"] / org_data["Air temperature [K]"]

org_data["product_id_num"] = pd.to_numeric(org_data["Product ID"].str.slice(start=1))
org_data.drop(columns='Product ID', inplace=True)
# we have transformed the product ID in str format using pandas library
```
